### delicious food combination

```python
bread = ["Rye bread", "Wheat", "White"]
meat = ['Meatball', 'Sausages', 'Chicken breast']
veggies = ['Lettuce', 'Tomato', 'Cucumber']
sauces = ['Mayonaise', 'Honey mustard', 'Chilli sauce']

print("Possible combinations of David's sandwich shops:")
for b in bread:
    for m in meat:
        for v in veggies:
            for s in sauces:
                print(f"{b} + {m} + {v} + {s}")
```
