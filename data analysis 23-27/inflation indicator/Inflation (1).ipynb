```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Load data

```python
cpi_ls_fe = pd.read_csv('CPILFESL.csv',index_col='DATE', parse_dates=True)  # Consumer Price Index Less Food and Energy
eci = pd.read_csv('ECIWAG.csv',index_col='DATE', parse_dates=True)          # Employment Price Index
ppi = pd.read_csv('PPIACO.csv', index_col='DATE', parse_dates=True)         # Producer Price Index
m2 = pd.read_csv('M2SL.csv',index_col='DATE', parse_dates=True)             # Money Supply
pce = pd.read_csv('PCE.csv',index_col='DATE', parse_dates=True)             # Personal Consumption Expenditure
```

```python
data = pd.concat([cpi_ls_fe, eci, ppi, m2, pce], axis=1)
```

```python
data = data.apply(pd.to_numeric, errors='coerce')
data_2018 = data['2018':]
```

```python
data_2018.columns = ['CPI', 'ECI', 'PPI', 'M2SL', 'PCE']
```

```python
data_quarterly = data_2018.resample('QS').mean().dropna()
print(data_quarterly.head(3))
print(f'Number of NaN : {data_quarterly.isna().sum().sum()}')
```

```python
data_quarterly[['ECI', 'PPI', 'CPI']].plot(figsize=(15, 5), xlabel='Year/Quaterly', ylabel='Index', grid=True)
plt.title('ECI, PPI, CPI')
plt.legend(title='Source: U.S. Bureau of Economic Analysis')
plt.show()
```

```python
data_quarterly[['M2SL', 'PCE']].plot(figsize=(15,5), xlabel='Year/Quaterly', ylabel='Dollar', grid=True)
plt.title('M2 & PCE')
plt.legend(title='Source: U.S. Bureau of Economic Analysis')
plt.show()
```
