```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import pearsonr

# Load the data
cpi_data = pd.read_csv('CPIAUCSL.csv', parse_dates=['DATE'], index_col='DATE')
wages_data = pd.read_csv('CES0500000003.csv', parse_dates=['DATE'], index_col='DATE')

# Preprocess the data
cpi_data.columns = ['CPI']
wages_data.columns = ['Wages']

# Combine the data into one DataFrame
data = pd.merge(cpi_data, wages_data, left_index=True, right_index=True)

# Resample to monthly frequency and calculate percentage changes
data = data.resample('M').ffill().pct_change().dropna()

# Perform correlation analysis
correlation, p_value = pearsonr(data['CPI'], data['Wages'])

# Perform linear regression
X = data['Wages']
y = data['CPI']
X = sm.add_constant(X)  # Add a constant term for the intercept
model = sm.OLS(y, X).fit()
regression_results = model.summary()

# Plot the data
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.plot(data.index, data['CPI'], label='CPI (Percent Change)')
plt.plot(data.index, data['Wages'], label='Wages (Percent Change)')
plt.title('CPI and Wage Growth (Percent Change)')
plt.legend()

plt.subplot(2, 1, 2)
plt.scatter(data['Wages'], data['CPI'])
plt.xlabel('Wage Growth (Percent Change)')
plt.ylabel('CPI (Percent Change)')
plt.title('Correlation between Wage Growth and CPI')

plt.tight_layout()
plt.show()

# Print results
print(f"Correlation between CPI and Wage Growth: {correlation:.2f} (p-value: {p_value:.4f})")
print(regression_results)
```

## correlation by converting to %

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess average hourly wage data
avg_hr_wage = pd.read_csv('CES0500000003.csv')
avg_hr_wage['DATE'] = pd.to_datetime(avg_hr_wage['DATE'])
avg_hr_wage = avg_hr_wage.set_index('DATE')
avg_hr_wage = avg_hr_wage['2006-03-01':]  # Ensure the same date range as CPI

# Calculate percentage change
avg_hr_wage['Wage_Pct_Change'] = avg_hr_wage['CES0500000003'].pct_change()

# Load and preprocess CPI data
cpi = pd.read_csv('CPIAUCSL.csv')
cpi['DATE'] = pd.to_datetime(cpi['DATE'])
cpi = cpi.set_index('DATE')
cpi = cpi['2006-03-01':]  # Ensure the same date range as wages

# Calculate percentage change
cpi['CPI_Pct_Change'] = cpi['CPIAUCSL'].pct_change()

# Plot the data
plt.figure(figsize=(20, 8))
plt.plot(avg_hr_wage.index, avg_hr_wage['Wage_Pct_Change'], label='Wage Percent Change')
plt.plot(cpi.index, cpi['CPI_Pct_Change'], label='CPI Percent Change')
plt.xlabel('Year')
plt.ylabel('Percent Change')
plt.title('Percentage Change of Average Hourly Wage and CPI')
plt.legend()
plt.show()

# Scatter plot to visualize the relationship
plt.figure(figsize=(20, 8))
plt.scatter(avg_hr_wage['Wage_Pct_Change'], cpi['CPI_Pct_Change'])
plt.xlabel('Wage Percent Change')
plt.ylabel('CPI Percent Change')
plt.title('Scatter Plot of Wage Percent Change vs. CPI Percent Change')
plt.show()

# Calculate and print the correlation
correlation = avg_hr_wage['Wage_Pct_Change'].corr(cpi['CPI_Pct_Change'])
print(f'Correlation between Wage Percent Change and CPI Percent Change: {correlation}')
```

## prediction using moving average

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
avg_hr_wage = pd.read_csv('CES0500000003.csv')
avg_hr_wage['DATE'] = pd.to_datetime(avg_hr_wage['DATE'])
avg_hr_wage = avg_hr_wage.set_index('DATE')
avg_hr_wage = avg_hr_wage['2006-03-01':]

cpi = pd.read_csv('CPIAUCSL.csv')
cpi['DATE'] = pd.to_datetime(cpi['DATE'])
cpi = cpi.set_index('DATE')
cpi = cpi['2006-03-01':]

# Calculate percentage change
avg_hr_wage['Wage_Pct_Change'] = avg_hr_wage['CES0500000003'].pct_change()
cpi['CPI_Pct_Change'] = cpi['CPIAUCSL'].pct_change()

# Calculate moving averages
avg_hr_wage['Wage_MA'] = avg_hr_wage['Wage_Pct_Change'].rolling(window=12).mean()
cpi['CPI_MA'] = cpi['CPI_Pct_Change'].rolling(window=12).mean()

# Plot the moving averages
plt.figure(figsize=(15, 10))
plt.plot(avg_hr_wage.index, avg_hr_wage['Wage_MA'], label='Wage Percent Change (MA)')
plt.plot(cpi.index, cpi['CPI_MA'], label='CPI Percent Change (MA)')
plt.xlabel('Year')
plt.ylabel('Percent Change (MA)')
plt.title('Moving Averages of Wage and CPI Percent Changes')
plt.legend()
plt.show()

# Forecasting future values (simple example)
forecast_period = 12  # Forecasting for the next 12 months
wage_forecast = avg_hr_wage['Wage_MA'].iloc[-1] * (1 + avg_hr_wage['Wage_Pct_Change'].mean()) ** forecast_period
cpi_forecast = cpi['CPI_MA'].iloc[-1] * (1 + cpi['CPI_Pct_Change'].mean()) ** forecast_period

print(f'Wage Forecast for next {forecast_period} months: {wage_forecast}')
print(f'CPI Forecast for next {forecast_period} months: {cpi_forecast}')
```
