# creating a dataframe

```python
# from dictionary
import pandas as pd
data = {'col_1': [3,2,1,0], 'col_2': ['a','b','c','d']}
pd.DataFrame(data)
```

```python
# from list
import pandas as pd
s = pd.Series([1.0, 2.0, 3.0], index = ['a', 'b', 'c'])
s2 = pd.Series([1.0, 2.0, 3.0, 4.0], index = ['a', 'b', 'c', 'd'])

data = {'one': s, 'two': s2}

pd.DataFrame(data)
```

# changing names of rows and columns

```python
# change the entire rows or columns name
import pandas as pd 
d = {'col1': [1,2], 'col2': [3,4]}
df = pd.DataFrame(data = d)
print(df)

print()

df.index = ['new index0', 'new index1'] # change index (row) names
df.columns = ['new col1', 'new col2']   # change feature (columns) names
print(df)
```

```python
# change selected rows or columns name
import pandas as pd
d = {'col1': [1,2], 'col2': [3,4]}
df = pd.DataFrame(data = d)
print(df)

pfchange = df.rename(index = {0: 'new index0', 1: 'new index2'}, columns = {'col2': 'new col2'})
print(pfchange)
```

```python
# change selected rows or columns name (with inplace = True)
import pandas as pd
d = {'col1': [1,2], 'col2': [3,4]}
df = pd.DataFrame(data = d)
df.rename(index = {0: 'new index0', 1: 'new index2'}, inplace = True)
df.rename(columns = {'col2': 'new col2'})
print(df)
```

# streaming pop song data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

song = pd.read_csv('/Users/anbschool0032/Desktop/data analysis/dataset/spotify_global_2019_most_streamed_tracks_audio_features.csv')

# create a new dataframe called df containing only the essentail columns
df = song[['Rank', 'Artist', 'Track Name', 'Artist_popularity', 'Streams', 'tempo', 'danceability', 'energy']]
print(df.head())

# create a new dataframe called pop_song by selecting df 'Artist_popularity' column, and set the condition to greater than 90 to filter
pop_song = df[df['Artist_popularity'] > 90]

# count the number of occurance of each artist that have the popularity of 90 or greater
rank = pop_song['Artist'].value_counts()

print('\n <Number of popular sings>')
print(rank)

# create new dataframes x and y to make a scatter plot graph for tempo and Rank columns
x = pop_song['tempo']
y = pop_song['Rank']
plt.scatter(x, y, color = 'red', alpha = 0.5)
plt.show()

# calculate the correlation coeffient
cor = pop_song.corr(method = 'pearson', numeric_only = True)
# create a matrix image 
plt.matshow(cor)
plt.colorbar

# sizing
plt.figure(figsize = [8, 6])
# format the image as a heatmap. using the 'annot=True' option, display the corresponding data values in each cell of the heatmap
sns.heatmap(cor, annot = True)
plt.show()

# create bts dataframe with the codition of Artist columns == BTS string for fav artist
bts = pop_song[pop_song['Artist'] == 'BTS']
print('\n <Track list for your favorite artist>')
print(bts)
```
