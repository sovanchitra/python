# Series

```python
import pandas as pd
s = pd.Series([1,2,3,4,5])
print(s)
```

```python
# create series from python lists or dictionaries

# from list
import pandas as pd
marvel = ['Iron Man', 'Captain America', 'Thor', 'Winter Soldier', 'Ultron', 'Ant-Man', 'Spider-man']
s = pd.Series(marvel)
print(s)

print()

# from dictionary
import pandas as pd
marvel_year = {'Iron Man': 2010, 'Captain America': 2011, 'Thor': '2013', 'Winter Soldier': 2014, 'Ultron': 2015}
s = pd.Series(marvel_year)
print(s)
s.index
```

```python
# create series from arrays generated by NumPy

# from np arange function
import pandas as pd
import numpy as np

n = np.arange(10, 16)
s = pd.Series(n)
print(s)

print()

# from random module
import pandas as pd
import numpy as np

np.random.seed(0)
r = np.random.random(size = 6)
s = pd.Series(r)
print(s)
```

```python
# create using values
import pandas as pd
import numpy as np

s = pd.Series(4)
print(s)
```

## one step further

```python
import pandas as pd
s = pd.Series([1,2,3,4,5], index = ['q', 1, 'cat', 'd', 'dog'])
# we can assign index to series
print(s)
```

```python
import pandas as pd
# create a list of dates
dates = pd.date_range('2021-01-01', '2021-01-07')
print(dates)
print()

marvel = ['Iron Man', 'Captain America', 'Thor', 'Winter Soldier', 'Ultron', 'Ant-Man', 'Spider-man']
# use the marvel list to create a series with the index of date list
new_s = pd.Series(marvel, index = dates)
print(new_s)
```

### selecting series element

```python
import pandas as pd

s = ['element1', 'element2', 'element3','element4', 'element5', 'element6', 'element7', 'element8']
sr = pd.Series(s)
sr[0]
sr[[4, 5, 2]]
```

```python
import pandas as pd

s = ['element1', 'element2', 'element3','element4', 'element5', 'element6', 'element7', 'element8']
sr = pd.Series(s, index = ('a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'))

sr['c']
sr[['a', 'h']]
```

# basic arithmetic operations on series

```python
import pandas as pd

s = [10, 20, 30, 40, 50]
sr = pd.Series(s)

print(sr)
print(sr ** 2)
```

# arithmetic operations on two or more series objects

```python
import pandas as pd
s = [10, 20, 30, 40, 50]
t = [23, 345, 123]
sr1 = pd.Series(s)
sr2 = pd.Series(t)
sr3 = sr1 + sr2
print(sr3)
```

# pandas provides many easy-to-use descriptive statistical functions

```python
import pandas as pd
s = [10, 20, 30, 40, 50]
t = [23, 345, 123]
sr1 = pd.Series(s)
sr2 = pd.Series(t)
sr3 = sr1 + sr2
print(sr3)
print(sr3.count())
print(sr3.sum())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

population_2020 = {'Tokyo': 37339804,
                    'Delhi': 31181376,
                    'Shanghai': 27795702,
                    'Sao Paulo': 22237472,
                    'Mexico City': 21918936,
                    'Dhaka': 21741090,
                    'Cairo': 21322750,
                    'Karachi': 16459472,
                    'Istanbul': 15415197,
                    'Buenos Aires': 15257673, 'Kinshasa': 14970460,
                    'Lagos': 14862111,
                    'Manila': 14158573,
                    'Rio de Janeiro': 13544462,
                    'Moscow': 12593252,
                    'Bogota': 11167392,
                    'Paris': 11078546,
                    'Jakarta': 10915364,
                    'Lima': 10882757}

population_2021 = {'Tokyo': 37393128,
                   'Delhi': 30290936,
                   "Shanghai": 27058480,
                   'Sao Paulo': 22043028,
                   'Mexico City': 21782378,
                   'Dhaka': 21005860,
                   'Cairo': 20900604,
                   'Karachi': 16093786,
                   'Istanbul': 15190336,
                   'Buenos Aires': 15153729,
                   'Kinshasa': 14342439,
                   "Lagos": 14368332,
                   'Manila': 13923452,
                   'Rio de Janeiro': 13458075,
                   "Moscow": 12537954,
                   'Bogota': 10978360, 
                   'Paris': 11017230,
                   'Jakarta': 10770487,
                   'Lima': 10719188}

city_area = {'Tokyo': 2194,
             'Delhi' :1484,
             'Shanghai': 6340,
            'Sao Paulo': 1521,
            'Mexico City': 1485,
            'Dhaka': 306.4,
            'Cairo': 3085,
            'Karachi': 3780,
            'Istanbul': 5343,
            'Buenos Aires': 203,
            'Kinshasa': 9965,
            'Lagos': 1171,
            'Manila': 42.88,
            'Rio de Janeiro': 1255,
            'Moscow': 2511,
            'Bogota': 1775,
            'Paris': 105.4,
            'Jakarta': 661.5,
            'Lima': 2672}

# convert to series
s_2020 = pd.Series(population_2020)
s_2021 = pd.Series(population_2021)
s_area = pd.Series(city_area)

# fluctuation calculation
growth = s_2020 - s_2021
print('<Increase or Decrease in the population>')
print(growth)

# horizontal bar graph
x = growth.index
y = growth.values
plt.figure(figsize = (13, 6))
plt.xticks(rotation = 'vertical', size = 10)
plt.barh(x, y)
plt.show()

# density calculation
population_density = s_2020 / s_area
print('<Population Density>')
print(population_density)
x = population_density.index
y = population_density.values
plt.figure(figsize = (13, 6))
plt.xticks(rotation = 45, size = 10)
plt.barh(x, y, color='red')
plt.title('population density', size = 15)
plt.ylabel('city', size = 15)
plt.xlabel('population', size = 15)
plt.show()

print('Minimum Population Density:', population_density.min())
print('Maximum Population Density:', population_density.max())
```
