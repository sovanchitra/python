```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

investment_dt = pd.read_csv('./private-investment-in-artificial-intelligence.csv')
investment_dt.head(15)
```

```python
investment_dt.describe()
```

```python
print('Remove Code column')
investment_dt = investment_dt.dropna(axis=1)
investment_dt
```

```python
investment_dt['Entity'].value_counts(dropna=True)
```

```python
investment_dt['Year'] = pd.to_datetime(investment_dt['Year'], format='%Y')
investment_dt
```

```python
investment_yr = investment_dt.set_index(investment_dt['Year'])
investment_yr
```

```python
total_investment = investment_yr[investment_yr['Entity'] == 'Total']
total_investment
```

```python
plt.plot(total_investment['Year'], total_investment['European Union and United Kingdom'])
plt.plot(total_investment['Year'], total_investment['China'])
plt.plot(total_investment['Year'], total_investment['United States'])
plt.plot(total_investment['Year'], total_investment['World'])
plt.show()
```

```python
plt.plot(total_investment['Year'], total_investment[total_investment['European Union and United Kingdom'] == ])
plt.plot(total_investment['Year'], total_investment[total_investment['China'] >= 1000000000 and total_investment['China'] >= 1000000000000])
plt.plot(total_investment['Year'], total_investment[total_investment['United States'] >= 1000000000 and total_investment['United States'] >= 1000000000000])
plt.plot(total_investment['Year'], total_investment[total_investment['World'] >= 1000000000 and total_investment['World'] >= 1000000000000])
plt.show()
```
