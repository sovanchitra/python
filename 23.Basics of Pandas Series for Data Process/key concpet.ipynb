### series. one dimensional array-like data

```python
import pandas as pd
s = pd.Series([1, 2 ,3, 4, 5])
print(s)   # index then data
```

#### create series from list

```python
import pandas as pd

marvel = ['Iron Man', 'Captain America', 'Thor', 'Winter Soldier', 'Ultron', 'Ant-Man', 'Spider-Man']

s = pd.Series(marvel)

print(s)
```

#### create series from dictionary

```python
import pandas as pd

marvel_year = {'Iron Man': 2010,
               'Captain America': 2011,
               'Thor': 2013, 
               'Winter Soldier': 2014,
               'Ultron': 2015}

s = pd.Series(marvel_year)

print(s)
s.index
```

#### create a series from arrays generated by NumPy

```python
import pandas as pd
import numpy as np

n = np.arange(10, 16)  # arange function generate integer 10-15

s = pd.Series(n)  # convert the array generated by NumPy into series object

print(s)
```

```python
import pandas as pd
import numpy as np

np.random.seed(0)

r = np.random.random(size = 6) # generate 6 random numbers from a normal distribution

s = pd.Series(r)

print(s)
```

#### create a series using values

```python
import pandas as pd
import numpy as np

s = pd.Series(4)

print(s)  # single index with value of 4
```

```python
import pandas as pd
s = pd.Series([1, 2, 3, 4, 5], index = ['q', 1, 'cat', 'd', 'dog'])
print(s)
```

```python
import pandas as pd

dates = pd.date_range('2021-01-01', '2021-01-07') # date range, (start date, end date)
print(dates)
```

```python
marvel = ['Iron Man', 'Captain America', 'Thor', 'Winter Soldier', 'Ultron', 'Ant-Man', 'Spider-Man']

new_s = pd.Series(marvel, index = dates)
print(new_s)
```

### selecting specific data(element)

```python
import pandas as pd

marvel = ['Iron Man', 'Captain America', 'Thor', 'Winter Soldier', 'Ultron', 'Ant-Man', 'Spider-Man']
s = pd.Series(marvel)
s.index  # can view the index values taht the series has
```

```python
s.values  # can view all data element
```

```python
import pandas as pd

s = ['lelement1', 'element2', 'element3','element4', 'element5','element6', 'element7', 'element8']
sr = pd.Series(s)

sr[2]
sr[[4, 5]]
```

```python
import pandas as pd

s = ['lelement1', 'element2', 'element3','element4', 'element5','element6', 'element7', 'element8']
sr = pd.Series(s, index = ('a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'))

sr['c']
sr[['e', 'f']]
```

### arithmetic operations on Series object

```python
import pandas as pd

s = [10, 20, 30, 40, 50]
sr = pd.Series(s)

print(sr)
print(sr ** 2)
```

```python
# operation on two or more series objects
# based on index
import pandas as pd

s = [10, 20, 30, 40, 50]
t = [23, 345, 123]
sr1 = pd.Series(s)
sr2 = pd.Series(t)

sr3 = sr1 + sr2
print(sr3)
```

### descriptive statistical functions

```python
import pandas as pd

s = [10, 20, 30, 40, 50]
t = [23, 345, 123]
sr1 = pd.Series(s)
sr2 = pd.Series(t)
sr3 = sr1 + sr2

print(sr3)
print(sr3.count())
print(sr3.sum())
```

### Matplotlib, a data visualization tool

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])  # y values
plt.show()
# automatically generate x values
```

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])  # (x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = range(100)
y = [value ** 2 for value in x]

plt.plot(x, y, linewidth = 5.0, color = 'red')

plt.title('hello world of chart')
plt.ylabel('y-some numbers')
plt.xlabel('x-some numbers')

plt.show
```

#### bar graph
