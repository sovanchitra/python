```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

population_2020 = {'Tokyo': 37339804,
'Delhi': 31181376,
'Shanghai': 27795702,
'Sao Paulo': 22237472,
'Mexico City': 21918936,
'Dhaka': 21741090,
'Cairo': 21322750,
'Karachi': 16459472,
'Istanbul': 15415197,
'Buenos Aires': 15257673,
'Kinshasa': 14970460,
'Lagos': 14862111,
'Manila': 14158573,
'Rio de Janeiro': 13544462,
'Moscow': 12593252,
'Bogota': 11167392,
'Paris': 11078546,
'Jakarta': 10915364,
'Lima': 10882757}


population_2021 = {'Tokyo': 37393128,
                   'Delhi': 30290936,
                   "Shanghai": 27058480,
                   'Sao Paulo': 22043028,
                   'Mexico City': 21782378,
                   'Dhaka': 21005860,
                   'Cairo': 20900604,
                   'Karachi': 16093786,
                   'Istanbul': 15190336,
                   'Buenos Aires': 15153729,
                   'Kinshasa': 14342439,
                   "Lagos": 14368332,
                   'Manila': 13923452,
                   'Rio de Janeiro': 13458075,
                   "Moscow": 12537954,
                   'Bogota': 10978360, 
                   'Paris': 11017230,
                   'Jakarta': 10770487,
                   'Lima': 10719188}

city_area = {'Tokyo': 2194,
             'Delhi' :1484,
             'Shanghai': 6340,
'Sao Paulo': 1521,
'Mexico City': 1485,
'Dhaka': 306.4,
'Cairo': 3085,
'Karachi': 3780,
'Istanbul': 5343,
'Buenos Aires': 203,
'Kinshasa': 9965,
'Lagos': 1171,
'Manila': 42.88,
'Rio de Janeiro': 1255,
'Moscow': 2511,
'Bogota': 1775,
'Paris': 105.4,
'Jakarta': 661.5,
'Lima': 2672}


s_2020 = pd.Series(population_2020)
s_2021 = pd.Series(population_2021)
s_area = pd.Series(city_area)

growth = s_2020 - s_2021
print('<Increase or Decrease in the populution>')
print(growth)

x = growth.index
y = growth.values
plt.figure(figsize = (13, 6))
plt.xticks(rotation = 'vertical', size = 10)
plt.barh(x, y)
plt.show()

population_density = s_2020 / s_area
print('<Population Density>')
print(population_density)

x = population_density.index
y = population_density.values
plt.figure(figsize = (13, 6))
plt.xticks(rotation = 45, size = 10)
plt.barh(x, y, color = 'red')
plt.ylabel('city', size = 15)
plt.xlabel('population', size = 15)
plt.show()

print('Minimum population density:', population_density.min())
print('Maximum population density:', population_density.max())
```
