```python
class BankAccount:

    def __init__(self, ID, name, deposit, withdraw1, withdraw2):
        """Initializes a BankAccount object with the given attributes.

        Args:
            ID: The account ID (string).
            name: The account owner's name (string).
            deposit: The initial deposit amount (float).
            withdraw1: The first withdrawal amount (float).
            withdraw2: The second withdrawal amount (float).

        Raises:
            ValueError: If `deposit`, `withdraw1`, or `withdraw2` is negative.
        """

        self.ID = ID
        self.name = name
        self.balance = deposit  # Initialize balance using the deposit amount

        # Input validation for non-negative values
        if deposit < 0 or withdraw1 < 0 or withdraw2 < 0:
            raise ValueError("Deposit, withdraw1, and withdraw2 must be non-negative.")

    def deposit(self, amount):
        """Deposits the given amount into the account.

        Args:
            amount: The amount to deposit (float).

        Returns:
            The new account balance (float).
        """

        if amount < 0:
            raise ValueError("Deposit amount cannot be negative.")

        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        """Withdraws the given amount from the account, if sufficient funds exist.

        Args:
            amount: The amount to withdraw (float).

        Returns:
            The new account balance (float), or None if withdrawal failed.

        Raises:
            ValueError: If the withdrawal amount is negative.
            RuntimeError: If the withdrawal amount exceeds the current balance.
        """

        if amount < 0:
            raise ValueError("Withdrawal amount cannot be negative.")

        if amount > self.balance:
            raise RuntimeError("Insufficient funds for withdrawal.")

        self.balance -= amount
        return self.balance

    def __str__(self):
        """Returns a formatted string representation of the bank account."""

        return f"Account Summary:\n" \
               f"  ID: {self.ID}\n" \
               f"  Name: {self.name}\n" \
               f"  Current Balance: {self.balance:.2f} won"

# Example usage
account1 = BankAccount('1231-0001', 'David', 2000, 500, 5000)

# Deposit 1000 won
new_balance = account1.deposit(1000)
print(f"New balance after deposit: {new_balance:.2f} won")

# Try to withdraw 6000 won (insufficient funds)
try:
    withdrawn_amount = account1.withdraw(6000)
    print(f"Withdrawn amount: {withdrawn_amount:.2f} won")
except RuntimeError as e:
    print(f"Withdrawal failed: {e}")

print(account1)  # Print the updated account summary
```
