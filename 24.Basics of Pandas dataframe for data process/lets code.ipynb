### data analysis of streaming popular songs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

song = pd.read_csv('spotify_global_2019_most_streamed_tracks_audio_features.csv')
df = song[['Rank', 'Artist', 'Track Name', 'Artist_popularity', 'Streams', 'tempo', \
          'danceability', 'energy']]
print(df.head())

pop_song = df[df['Artist_popularity'] > 90]
rank = pop_song['Artist'].value_counts()
print('\n <Number of popular songs>')
print(rank)

x = pop_song['tempo']
y = pop_song['Rank']
plt.scatter(x, y, color = 'red', alpha = 0.5)
plt.show()

cor = pop_song.corr(method = 'pearson', numeric_only = True)
plt.matshow(cor)
plt.colorbar

plt.figure(figsize = (8, 6))
sns.heatmap(cor, annot = True)
plt.show()

bts = pop_song[pop_song['Artist'] == 'BTS']
print('\n <Track list for your favorite artist>')
print(bts)
```
